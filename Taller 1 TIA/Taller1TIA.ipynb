```python
%matplotlib inline
import cv2
from IPython.display import Image
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
img1 = cv2.imread('fotomano/interior.jpg',0) # Carga imagen en grises
img2 = cv2.imread('fotomano/soleado.jpg',0) 
img3 = cv2.imread('fotomano/sombra.jpg',0)
```

```python
cv2.imshow('Interior',img1)
cv2.imshow('Soleado',img2)
cv2.imshow('Sombra',img3)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

## Histograma imagen interior de la mano
* Histograma con 8 
* Histograma con 32
* Histograma con 256 bits

```python

hist10 = cv2.calcHist([img1],[0],None,[8],[0,256])
hist11 = cv2.calcHist([img1],[0],None,[32],[0,256])
hist12 = cv2.calcHist([img1],[0],None,[256],[0,256])

plt.figure(1)

plt.subplot(221), plt.imshow(img1, cmap='gray'), plt.title('Imagen')
plt.subplot(222), plt.hist(img1.ravel(),8,[0,256]) ,plt.title('8 Bins')
plt.subplot(223), plt.hist(img1.ravel(),32,[0,256],facecolor='r'), plt.title('32 Bins')
plt.subplot(224), plt.hist(img1.ravel(),256,[0,256],facecolor='g'), plt.title('256 Bins')

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```
