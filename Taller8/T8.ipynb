# Ejemplo de estudio "Fuzzy Control Systems: The Tipping Problem"

En este caso la aplicación principal del sistema de control de lógica difusa se centra en determinar qué tanta propina se debería dejar en un servicio de restaurante entre 0% y 25% de lo consumido, para ello se definen las entradas y salidas del sistema, en este caso las entradas son dos elementos llamados "servicio" y "calidad de la comida" ambos ámbitos calificados entre 0 y 10.

Los antecedentes serían entonces:

- Servicio
 - Universo de 0 a 10, ¿Qué tan bueno fue atendido?
 - Conjunto difuso: inaceptable, aceptable y bueno
- Calidad de la comida
 - Universo de 0 a 10
 - Conjunto: malo, decente, bueno

El consecuente o salida será finalmente:

- Propina
 - Universo, ¿De cuánto será la propina del 0% al 25%?
 - Conjunto difuso: bajo, medio, alto

Las reglas para hallar la respuesta serán entonces:

- SI el servicio fue bueno o la calidad de la comida fue bueno, ENTONCES la propina será alta.
- SI el servicio fue aceptable, ENTONCES la propima será media.
- SI el servicio era inaceptable y la calidad de la comida era mala ENTONCES la propina será baja.

Los siguientes pasos ya serán la implementación del algoritmo difuso y su codificación en python que será comentada

```python
#Importación de librerías.
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Se definen las variables con su titulo y universo
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')


#Se definen cómo se van a separar los intervalos entre el conjunto difuso, se puede hacer de forma automatica o manual como se muestra abajo:

# Auto-membership function population is possible with .automf(3, 5, or 7)
quality.automf(3)
service.automf(3)

# Custom membership functions can be built interactively with a familiar,
# Pythonic API
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 12.5])
tip['medium'] = fuzz.trimf(tip.universe, [0, 12.5, 25])
tip['high'] = fuzz.trimf(tip.universe, [12.5, 25, 25])
```

```python
#Se muestran finalmente las graficas de lógica difusa que se usarán
# You can see how these look with .view()
quality['average'].view()
```

```python
service.view()
```

```python
tip.view()
```

```python
#Se deginen las reglas cómo se habían descrito anteriormente, notese el operador or "|"

rule1 = ctrl.Rule(quality['poor'] | service['poor'], tip['low'])
rule2 = ctrl.Rule(service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] | quality['good'], tip['high'])

rule1.view()
```

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3]) 
#Esta función genera un sistema en el cual se aplican las reglas definidas anteriormente, es diferente crear el sistema a simularlo.    
```

```python
tipping = ctrl.ControlSystemSimulation(tipping_ctrl) # En este caso se simula el sistema, es decir, se genera un objeto el cual recibirá las entradas
#numericas que se aplicarán al sistema creado anteriormente para un caso especifico
```

```python
#Se ingresan las entradas númericas a cada antecedente

# Pass inputs to the ControlSystem using Antecedent labels with Pythonic API
# Note: if you like passing many inputs all at once, use .inputs(dict_of_data)
tipping.input['quality'] = 6.5
tipping.input['service'] = 9.8

#Finalmente se calcula la salida del sistema de lógica difusa 
# Crunch the numbers
tipping.compute()
```

```python
#Se imprime la salida obtenida
print (tipping.output['tip'])
tip.view(sim=tipping)
```

En este caso finalmente se obtiene que la propina para un servicio de 6.5 y una comida de 9.8 es de 19.76% de lo consumido
